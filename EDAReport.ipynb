## Introduction

### 1.1 Background & Objective
The goal of this project is to analyze customer credit data to identify factors that influence credit risk.
This can help financial institutions make better lending decisions.

### 1.2 Dataset Description
- Dataset: German Credit (OpenML)
- Observations: 1000
- Features: 20 independent variables capturing customer characteristics such as age, loan amount, duration, employment status, and financial history
- Target: credit_risk (binary outcome indicating whether a customer is classified as good or bad credit risk)

### 1.3 Key Questions
- What factors influence credit risk?
- Do higher loan amounts increase risk?
- Does loan duration affect default probability?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/selva86/datasets/master/GermanCredit.csv"
df = pd.read_csv(url)

df.head()
```

# 📌 Categorical Variables

```python
import seaborn as sns

cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    plt.figure(figsize=(8,4))

    counts = df[col].value_counts()

    sns.barplot(
        x=counts.values,
        y=counts.index,
        palette='coolwarm'
    )

    plt.title(f"{col} Distribution", fontsize=12)
    plt.xlabel("Count")
    plt.ylabel("")

    for i, v in enumerate(counts.values):
        plt.text(v + 5, i, str(v), va='center')

    plt.tight_layout()
    plt.show()
```

### Insight
Most categories are imbalanced, with a few dominant groups.
This may introduce bias into predictive models.

# 🤖 Machine Learning

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

df_ml = df.copy()

le = LabelEncoder()
for col in df_ml.select_dtypes(include='object').columns:
    df_ml[col] = le.fit_transform(df_ml[col])

X = df_ml.drop('credit_risk', axis=1)
y = df_ml['credit_risk']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

pred = model.predict(X_test)

print(classification_report(y_test, pred))
```

# Key Insights

1. Loan amount is a strong indicator of credit risk  
2. Longer loan duration increases default probability  
3. Dataset is imbalanced (more good customers)  
4. Age has weak predictive power  
5. Feature engineering improves model performance  

# Conclusion

This analysis shows that:

- Larger loan amounts increase credit risk  
- Longer durations are associated with higher default probability  
- Age has limited predictive power

Banks should carefully evaluate high-value and long-term loans.
