**Why the Memory Addresses are Different:**
- **Temporary Instance (id(Cheff())):**
> Each call to Cheff() creates a new, unique instance of the Cheff class.
When you call id(Cheff()), a new instance is created temporarily for the id function call, and this instance is not stored or referenced after the id call.
Since it is a new instance, it has a unique memory address each time you call Cheff().

- **Persistent Instance (id(user_1)):**
> When you call user_1 = Cheff(), an instance of Cheff is created and stored in the variable user_1.
The id(user_1) call retrieves the memory address of this specific, persistent instance.

```python
class Car:
    # 1. 생성자: 객체가 만들어질 때 호출되는 함수
    def __init__(self, brand, model, color):
        self.brand = brand  # 브랜드
        self.model = model  # 모델명
        self.color = color  # 색상
        self.speed = 0      # 현재 속도 (기본값 0)

    # 2. 메서드: 자동차의 동작 (가속)
    def accelerate(self, amount):
        self.speed += amount
        print(f"{self.model}이(가) {amount}km/h 가속하여 현재 {self.speed}km/h입니다.")

    # 3. 메서드: 자동차의 동작 (정지)
    def stop(self):
        self.speed = 0
        print(f"{self.model}이(가) 멈췄습니다.")
```

```python
# 'my_car'라는 실제 자동차 객체 생성
my_car = Car("Hyundai", "Ionic 6", "White")

# 'your_car'라는 또 다른 자동차 객체 생성
your_car = Car("Tesla", "Model 3", "Black")

# 각 자동차의 정보 출력
print(f"내 차는 {my_car.brand} {my_car.model}입니다.")
# 출력: 내 차는 Hyundai Ionic 6입니다.

# 자동차 동작시키기
my_car.accelerate(50)  # Ionic 6이 50km/h로 달립니다.
your_car.accelerate(80) # Model 3는 80km/h로 달립니다.
```

## **과제 1. faker를 사용하여 1000명의 데이터를 class를 이용하여 만들어라**

- [click faker](https://faker.readthedocs.io/en/master/)

```python
!pip install Faker
```

```python
from faker import Faker
import pandas as pd
import random

fake = Faker()

class User:
    def __init__(self, user_id, name, email, address, age):
        self.id = user_id
        self.name = name
        self.email = email
        self.address = address
        self.age = age

    def to_dict(self):
        return {
            "id": self.id,
            "name": self.name,
            "email": self.email,
            "address": self.address,
            "age": self.age
        }


users = []

for i in range(1000):
    user = User(
        user_id=str(i).zfill(4),
        name=fake.name(),
        email=fake.email(),
        address=fake.address(),
        age=random.randint(18, 70)
    )
    users.append(user)


df = pd.DataFrame([user.to_dict() for user in users])

print(df.head())
```

## **과제 2. Tom이 Jerry를 쫒아가는 행위를 method(함수)로 만들어 보자(10번 following)**

### **class User:**
> **def __init __ (self):**
>   - ##### Initialize attributes
>   - **HAS (Attiributes)**

> **def following_mode():**
>   - ##### change attributes (following, follower) for chasing
>   - **Does (Method, Change the attributes)**

<img src= 'https://pbs.twimg.com/profile_images/810136393/tom-jerry-pic_400x400.jpg' width=300>

```python
class User:
    def __init__(self, name):
        self.name = name
        self.following = None
        self.followers = []

    def following_mode(self, other_user):
        if self not in other_user.followers:
            self.following = other_user
            other_user.followers.append(self)

        print(f"{self.name} is chasing {other_user.name}")
```

```python
tom = User("Tom")
jerry = User("Jerry")
```

```python
for i in range(10):
    tom.following_mode(jerry)
```

```python
print("Tom is following:", tom.following.name)
print("Jerry followers:", [u.name for u in jerry.followers])
```

```python
import gradio as gr

class User:
    def __init__(self, name):
        self.name = name
        self.following = None
        self.followers = []

    def following_mode(self, other_user):
        if self not in other_user.followers:
            self.following = other_user
            other_user.followers.append(self)

        return f"{self.name} is chasing {other_user.name}"


# создаём объекты
tom = User("Tom")
jerry = User("Jerry")


# функция для Gradio
def chase():
    result = []
    for i in range(10):
        result.append(tom.following_mode(jerry))

    followers = [u.name for u in jerry.followers]

    return "\n".join(result) + f"\n\nJerry followers: {followers}"


# интерфейс
app = gr.Interface(
    fn=chase,
    inputs=[],
    outputs="text",
    title="Tom chases Jerry 🐱🐭"
)

app.launch()
```
